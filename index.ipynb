## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

SQL is a relational database, while NoSQL is not. NoSQL does not require a schema and can hold varying entry structures.

## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

It stores data as a key-value entry, like a JSON object. It's different from a column in a relational database because it can have nested levels of key-values.




## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Fault tolerance means that the server structure is robust such that if one server goes down others are there to back it up.

Resilient distributed datasets are fault tolerant because they incorporate a distributed architecture of servers that can back each other up if others go down.







## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MapReduce is a two step process used by schemaless databases to "map" data based on its structure and then reduce it to a smaller size.
